```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wiki = pd.read_csv("paths_finished.csv",sep=",")
wiki.head(20)
```

```python
wiki.dropna(axis=0,subset=['path'],inplace=True)
```

```python
wiki['path_list'] = wiki['path'].apply(lambda x: x.split(';'))
wiki.head()
```

```python
path = wiki['path_list']
freq_item_set_1 = {}
```

```python
def compute_occurrance(x):
    for item in x:
        freq_item_set_1[item] = freq_item_set_1.get(item,0) + 1
    
path.apply(compute_occurrance)
freq_item_set_1
```

```python
MIN_SUP = 2     
MIN_SUP
```

```python
F1 = [key for key,value in freq_item_set_1.items() if value > MIN_SUP]
patterns = set()
[patterns.add(cand) for cand in F1]
len(patterns)
```

```python
import itertools
permutations = itertools.permutations(patterns,2)
freq_item_set_2 = {}
# C2 = []

```

```python
for i,perm in enumerate(permutations):
        for row in path[:100]:
            if set(perm) <= set(row):
                print(i)
                freq_item_set_2[perm] = freq_item_set_2.get(perm,0) + 1
freq_item_set_2
```

```python
F2 = [key for key,value in freq_item_set_2.items() if value > MIN_SUP]
[patterns.add(cand) for cand in F2]
len(patterns)
```

```python
C3 = []
combinations = itertools.permutations(F2,2)
freq_item_set_3 = {}
for comb in combinations:
    if comb[0][1] == comb[1][0]:
        C3.append([comb[0][0],comb[0][1],comb[1][1]])
```
